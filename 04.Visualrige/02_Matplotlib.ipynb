### 08. Matplotlib 그래프 영역 채우기

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x[1:3], y[1:3], alpha=0.5) ## fill_between() 사용 alpha는 투명도를 의미
# plt.fill_betweenx(y[2:4], x[2:4], color='pink', alpha=0.5)  ## fill_betweenx() 사용
plt.axhline(0); plt.axvline(0) # 수평선 x축 y축
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_betweenx(y[2:4], x[2:4], color='pink', alpha=0.5)  ## fill_betweenx() 사용
plt.axhline(0); plt.axvline(0) # 수평선 x축 y축
plt.show()
```

### 두 그래프 사이 영역 채우기

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x[1:3], y1[1:3], y2[1:3], color='lightgray', alpha=0.5)

plt.show()
```

```python
x = np.arange(1.0, 4.0, 0.01)
y1 = x ** 2
y2 = (x ** 2) / 2

plt.plot(x, y1); plt.plot(x, y2)
plt.xlabel('X-Label'); plt.ylabel('Y-Label')
plt.title('두 곡선 사이의 영역 채우기')
plt.fill_between(x[100:200], y1[100:200], y2[100:200], color='lightgray', alpha=0.5)

plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.title('임의의 영역 채우기')
plt.fill([1.9, 1.9, 3.1, 3.1], [2, 5, 11, 8], color='lightgray', alpha=0.5)
# 주어진 점을 기준으로 영역을 채운다.
plt.show()
```

### 10. Matplotlib 그리드 설정하기

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True)

plt.show()
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y') # y축 기준으로 그어진 grid만 보여진다.
plt.title('Grid축 지정하기')
plt.show()
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y', color='red', alpha=0.5, linestyle='--')
plt.title('Grid스타일 설정하기')
plt.show()

# which 파라미터를 ‘major’, ‘minor’, ‘both’ 등으로 사용하면 주눈금, 보조눈금에 각각 그리드를 표시할 수 있습니다.
```

### 11. Matplotlib 눈금 표시하기

### 12. Matplotlib 타이틀 설정하기

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
plt.title('Sample graph', loc='right', pad=10)
# 제목도 위치 조절이 가능
plt.show()
```

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
plt.title('Sample graph', loc='right', pad=10)
title_font = {
    'fontsize': 16,
    'fontweight': 'bold',
    'color': 'magenta'
}
# 폰트 설정도 가능
plt.title('Sample graph', fontdict=title_font, loc='left', pad=20)
plt.show()
```

```python
x = np.linspace(-4.0, 4.0, 801)
nd = np.exp(-x*x/2)/np.sqrt(2*np.pi)
plt.plot(x, nd)
plt.axhline(0, alpha=0.5); plt.axvline(0, alpha=0.5)
plt.show()
```

```python
a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step', label='a')
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled', label='b')
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step', label='c')
plt.legend()
plt.show()

# label 설정 후 plt.legend를 하면 범례표가 완성
```

### 20. Matplotlib 파이 차트 그리기

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
# 반시계 방향으로 나열된다.
plt.pie(ratio, labels=labels, autopct='%.1f%%')
# autopct는 부채꼴 안에 표시될 숫자의 형식을 지정한다. 소수점 한자리까지 표시하도록 설정한다.
plt.show()
```

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False)
# startangle는 부채꼴이 그려지는 시작 각도를 설정합니다. 디폴트는 0도 (양의 방향 x축)로 설정되어 있습니다. counterclock=False로 설정하면 시계 방향 순서로 부채꼴 영역이 표시됩니다.
plt.show()
```

### 중심에서 벗어나는 정도 설정하기

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```
