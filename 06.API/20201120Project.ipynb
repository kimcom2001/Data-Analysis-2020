```python
import pandas as pd
import folium 
import requests
from urllib.parse import urlparse
```

```python
df = pd.read_csv('커피업소정보.csv', encoding='euc-kr')
```

```python
print(df.columns)
```

```python
dataset = df[['상호명','시도명','시군구명', '위도', '경도']]

dataset.head()
```

```python
starbucks_list = dataset[(dataset['상호명']=='스타벅스')&(dataset['시도명']=='서울특별시')&(dataset['시군구명']=='강서구')]
print('강서구 내 전문점 점포 수 :', len(starbucks_list))
starbucks_list
```

```python
baek_list = dataset[(dataset['상호명']=='빽다방')&(dataset['시도명']=='서울특별시')&(dataset['시군구명']=='강서구')]
print('강서구 내 전문점 점포 수 :', len(baek_list))
baek_list
```

```python
twosome_list = dataset[(dataset['상호명']=='투썸플레이스')&(dataset['시도명']=='서울특별시')&(dataset['시군구명']=='강서구')]
print('강서구 내 전문점 점포 수 :', len(twosome_list))
twosome_list
```

```python
ediya_list = dataset[(dataset['상호명']=='이디야커피')&(dataset['시도명']=='서울특별시')&(dataset['시군구명']=='강서구')]
print('강서구 내 전문점 점포 수 :', len(ediya_list))
ediya_list
```

```python
x = pd.concat([starbucks_list, baek_list, twosome_list, ediya_list])
x.head()
```

```python
x.to_csv('coffee.csv',index=False)
```

```python
coffee_list = pd.read_csv('coffee.csv')
```

```python
coffee_list.head()
```

```python
# 투썸플레이스와 이디야 커피는 영어이름으로 저장되어 있는데 분류를 하지 못해 지도에 넣지 못함(자료 검색시 조심할 사항)

mapping = folium.Map(location=[coffee_list['위도'].mean(), coffee_list['경도'].mean()], zoom_start=14)

color_dict = {'스타벅스': 'blue', '이디야커피': 'green', '투썸플레이스': 'red', '빽다방': 'gray'}
icon_dict = {'스타벅스': 'glyphicon glyphicon-globe', '이디야커피': 'glyphicon glyphicon-unchecked', '투썸플레이스': 'glyphicon glyphicon-heart-empty', '빽다방': 'cloud'}

for i in coffee_list.index:
    folium.Marker(
    location=[coffee_list['위도'][i], coffee_list['경도'][i]],
    popup=coffee_list['상호명'][i],
    icon=folium.Icon(color=color_dict[coffee_list['상호명'][i]], icon= icon_dict[coffee_list['상호명'][i]])
).add_to(mapping)

folium.CircleMarker(
    location=[coffee_list['위도'].mean(), coffee_list['경도'].mean()],
    radius=500,
    popup='원 그림',
    color='red'
).add_to(mapping)

mapping
```
